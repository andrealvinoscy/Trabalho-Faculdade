```python
class Pessoa:
  def __init__(self,nome,idade):
    self.nome=nome
    self.idade=idade
  def apresentar(self):
    print(f'Olá, sou o funcionário(a) {self.nome} e minha idade é {self.idade}')

p1 = Pessoa('Mario', 30)
p1.apresentar()

p2 = Pessoa('Luigi', 25)
p2.apresentar()
```

```python
class Cliente(Pessoa):
  quantidadeClientes = 0
  def __init__(self, nome, idade):
    super().__init__(nome, idade)
    Cliente.quantidadeClientes+=1
    cadastro = 1000 + Cliente.quantidadeClientes
    self.cadastro = cadastro
  def apresentar(self):
    super().apresentar()
    print(f'Seu cadastro é {self.cadastro}')


c1 = Cliente('Toad', 15)
c1.apresentar()
print(Cliente.quantidadeClientes)

```

```python
class funcionario(Pessoa):
  #pass
  def __init__(self,nome,idade,cadastro):
    self.nome=nome
    self.idade=idade
    self.cadastro = cadastro
  def apresentar(self):
    print(f'Olá, sou o funcionario(a) {self.nome} e minha idade é {self.idade}')
f1 = funcionario('Peach', 20, 1001)
f2 = funcionario('Bowser', 42, 1002)
f1.apresentar()
f2.apresentar()
print(f1.cadastro, f2.cadastro)
```

```python
from ast import Add
class Aluno:
  totalAlunos = 0
  def __init__(self,nome,nota):
    self.nome=nome
    self.nota=nota
    Aluno.totalAlunos+=1
    self.nome=nome
    self.nota=nota
    self.ru=1000 +Aluno.totalAlunos
  def info(self):
    print(f'Nome: {self.nome},RU: {self.ru} Nota: {self.nota}')

  def getNota(self):
    return self.nota
a1 = Aluno('Andre', 92)
a1.info()

class Turma:
  def __init__(self,nome,limiteAlunos):
    self.nome=nome
    self.limiteAlunos=limiteAlunos
    self.listaAlunos=[]

  def addAluno(self,aluno):
    if len(self.listaAlunos)<self.limiteAlunos:
      self.listaAlunos.append(aluno)
      return True
    return False

  def mediaTurma(self):
    soma=0
    for aluno in self.listaAlunos:
      soma+=aluno.getNota()
    return soma/len(self.listaAlunos)


a1 = Aluno('Amanda', 91)
a2 = Aluno('Leticia', 87)
a3 = Aluno('Valeria', 100)
t1 = Turma('Minha Turma', 5)
t1.addAluno(a1)
t1.addAluno(a2)
t1.addAluno(a3)

print(t1.addAluno(a3))
```
